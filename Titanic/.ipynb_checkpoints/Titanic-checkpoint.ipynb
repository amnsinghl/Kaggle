```python
def readDataFromFile(filename):
    from numpy import genfromtxt
    print "reading data please wait..."
    my_data = genfromtxt(filename, delimiter=',',skip_header=1)
    print my_data.shape
    return my_data
```

```python
def readDataFromFile2(filename):
    import csv
    ret = []
    with open(filename, 'rb') as csvfile:
        spamreader = csv.reader(csvfile)
        for row in spamreader:
            ret.append(row)
        return ret
```

```python
def writeFile(y_data):
    f = open('myfile','w')
    f.write("PassengerId,Survived\n")
    i = 892;
    for k in y_data:
        f.write(str(i) + "," + str(int(k)) + "\n")
        i = i+1
    f.close()
```

```python
def checkIfNull(string):
    if len(str(string)) == 0:
        return True
    else:
        return False
    
def intToList(num, siz):
    ret = [0] * siz
    i=0
    while num > 0:
        ret[i] = num%2
        num = num/2
        i = i+1
    return ret

def tClass(row):
    num = int(row[2])
    return intToList(num,2)

def tAge(row):
#     15,30,45,60,75
    val = row[5]
    if checkIfNull(val):
        val = 0
    num = int(float(val)/15)
    return intToList(num, 4)

def tSex(row):
    if row[4] == 'male':
        return [0]
    else:
        return [1]
def tSb(row):
    num = int(row[6])
    return intToList(num,4)

def tSp(row):
    num = int(row[7])
    return intToList(num,4)

def tCost(row):
    val = row[9]
    if checkIfNull(val):
        val = 0
    num = int(round(float(val)))
    ret = 0
    while num > 0:
        num = num/2
        ret = ret + 1
    return intToList(ret, 8)

def tEm(row):
    val = 0
    if row[11] == 'S':
        val = 1
    elif row[11] == 'Q':
        val = 2
    elif row[11] == 'C':
        val = 3
    return intToList(val, 2)

def transform(row):
#     print row
    y = row[1]
    x = []
    for t in tClass(row):
        x.append(t)
    for t in tSex(row):
        x.append(t)
    for t in tAge(row):
        x.append(t)
    for t in tSb(row):
        x.append(t)
    for t in tSp(row):
        x.append(t)
    for t in tCost(row):
        x.append(t)
#     print x
    return [x, y]
```

```python
train_data = readDataFromFile2('train.csv')
```

```python
import numpy
flag = True
X = []
Y = []
for row in train_data:
    if flag:
        flag = False
        continue
    x, y = transform(row)
    X.append(numpy.array(x))
    Y.append(y)
X = numpy.array(X)
Y = numpy.array(Y)
print X.shape, Y.shape
```
