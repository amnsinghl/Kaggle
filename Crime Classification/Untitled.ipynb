```python
def convertToWeek(dd):
    for i in range(0,len(week)):
        if dd==week[i]:
            return i;
```

```python
def convertToTime(dd):
    t = time.strptime(dd, "%Y-%m-%d %H:%M:%S")
    return t.tm_hour
#     return float(t.tm_hour) +  float(t.tm_min)/60.0
```

```python
def convertToDate(dd):
    t = time.strptime(dd, "%Y-%m-%d %H:%M:%S")
    return t.tm_mday
```

```python
def convertToMonth(dd):
    t = time.strptime(dd, "%Y-%m-%d %H:%M:%S")
    return t.tm_mon
```
